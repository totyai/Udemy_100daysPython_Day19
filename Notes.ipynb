## Event Listeners
- These will notice if there is any keypress, that the user does on the keyboard.
- In Turtle we have a *listen* function that does this
- This function expects another funtion as argumnet, to tell what it does, when the key is being pressed.
- When we pass a function as argument, we **DON'T** write (), as it would trigger it.

```python
from turtle import Turtle, Screen

tim = Turtle()
screen = Screen()

def move_forwards():
    tim.forward(10)

screen.listen()
screen.onkey(key="space", fun=move_forwards)
screen.exitonclick()
```

## Higher order function
It is a function, that works with other fucntions as input.
So whent his higher order function is being created,, it expects as an input another function's name. Then it triggers the other fucntion inside of itself.

```python
def add(n1,n2): # These are the sub-funcitons
    return n1 + n2

def multiply(n1,n2): # These are the sub-functions
    return n1 * n2

def calculator(n1,n2,func): #This is the higher order funciton, as it get another as argument, without ()
    return func(n1,n2) # This is where the other function get triggered. We don't care about the name of the funciton originally

result = calculator(3,5, multiply) # This calls in the higher order function
print(result)
```
